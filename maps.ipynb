```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("model_predictions_by_community.csv")
com_areas = gpd.read_file("data/com_areas_chi")
```

```python
df = df.rename(columns={"community":"area_numbe"})
df["difference"] = df["Prediction LST"] - df["Actual LST"]
```

```python
com_areas["area_numbe"] = com_areas["area_numbe"].astype("int")
com_areas = com_areas.merge(df, on = "area_numbe")
```

```python
com_areas
```

```python
a = com_areas.plot("Prediction LST", legend = True)

plt.savefig('plots/prediction.jpg')
```

```python
b = com_areas.plot("Actual LST", legend = True)
plt.savefig('plots/actual.jpg')
```

```python
c = com_areas.plot("difference", legend = True, cmap="Blues")
plt.savefig('plots/difference.jpg')
```

```python
fig, axes = plt.subplots(ncols=3)
# add geopandas plot to left subplot
com_areas.plot("Prediction LST", legend = True, ax=axes[0])
com_areas.plot("Actual LST", legend = True, ax=axes[1])
com_areas.plot("difference", legend = True, ax=axes[2])

```
